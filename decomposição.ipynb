# Everything Plus — Decomposição

## Objetivo

Segmentar usuários com base em seus perfis de consumo, analisando o histórico de compras por tipo de produto, com o intuito de gerar insights para personalização de ofertas e estratégias de marketing.

---

## Etapas principais

### 1. Pré-processamento dos dados
- Carregar o dataset
- Tratar valores ausentes
- Corrigir ou remover valores inválidos
- Criar coluna com o total gasto por item

### 2. Análise exploratória de dados
- Criar categorias de produto com base na descrição
- Calcular, para cada cliente:
  - Total de compras por categoria
  - Frequência por categoria
  - Gasto total por categoria
- Criar uma linha por cliente com essas informações
- Observar padrões gerais de comportamento de compra

### 3. Clusterização e análise dos clusters
- Normalizar os dados, se necessário
- Definir o número ideal de clusters
- Atribuir cada cliente a um grupo
- Analisar o perfil de cada cluster com base nas métricas por categoria
- Gerar insights acionáveis

### 4. Testes de hipóteses
- Formular hipóteses para comparar os clusters
- Aplicar testes estatísticos 
- Interpretar os resultados e verificar a significância

### 5. Conclusões e apresentação

- Descrever o perfil de cada cluster e sugerir ações de marketing personalizadas para cada grupo
- Criar uma apresentação em PDF com:
  - Introdução
  - Metodologia e etapas do projeto
  - Resultados da segmentação
  - Conclusões e recomendações de negócio
- Adicionar o link da apresentação

### 6. Dashboard
- Desenvolver um dashboard no Tableau para exploração visual dos dados, contendo:
  - Total de compras por dia
  - Distribuição do total de compras por usuário
  - Filtro por data de compra
